```python
sns.boxplot(x='Churn', y='TotalCharges', data=data)
plt.title('Total Charges vs Churn')
plt.show()
```

1. The Box Plot (The "Timing" Clue)
Imagine you have two buckets of people: those who stayed (No) and those who quit (Yes).

The "Middle Line" (Median): Look at the horizontal line inside the blue boxes. In the Yes box, that line is much lower down the Y-axis.

The Logic: This tells us that half of the people who quit had spent less than ~$500 total. In contrast, the people who stay have a much higher median spending.

The Conclusion: Most people who quit are doing it very early. They aren't staying long enough to run up a big bill. If they were quitting because of long-term service issues, that box would be much higher up.

```python
sns.scatterplot(x='tenure', y='TotalCharges', hue='Churn', data=data, alpha=0.5)
```

2. The Scatter Plot (The "Life Story" Clue)
This plot shows every single customer as a dot. The X-axis (tenure) is "How many months have you been here?" and the Y-axis is "How much have you paid us in total?"

The "Fan" Shape: See how the dots spread out into a triangle?

The Top Edge: These are your "Premium" customers. They pay a lot every month, so their total bill goes up very fast (steep slope).

The Bottom Edge: These are your "Budget" customers. Even after 70 months, their total bill is still low because they pay very little per month.

The Colors (Churn): Notice the concentration of Orange dots (Yes) at the very bottom-left corner (0-10 months).

The Logic: It proves that Tenure is the biggest predictor of TotalCharges. You can’t have high TotalCharges without high Tenure.

The Conclusion: Since almost all the orange dots are at the "start" of the lines, it confirms that your churn problem is an onboarding problem. Once a customer survives past the 20-month mark (the middle of the plot), the dots turn mostly blue, meaning they are much more likely to stay forever.

```python
sns.kdeplot(data=data, x='MonthlyCharges', hue='Churn', shade=True)
```

KDE Plot: Monthly Charges vs. Churn
This graph shows the "density" or concentration of customers at different monthly price points.

The Blue Peak ($20): Look at that massive blue mountain at $20. Those are customers on basic/low-cost plans who rarely churn. They are very stable.

The Orange "Danger Zone" ($70–$110): Notice how the orange area (Churn: Yes) gets much taller between $70 and $110.

The Insight: High monthly charges are a major "trigger" for leaving. People are okay paying $20, but once the bill hits $80+, they start looking for the exit.

```python
 
# 1. Create a normalized crosstab (gives percentages per row)
# We multiply by 100 to get whole numbers for the labels
df_pct = pd.crosstab(data['SeniorCitizen'], data['Churn'], normalize='index') * 100

# 2. Plot
ax = df_pct.plot(kind='bar', stacked=True, color=['#3caea3', '#f6d55c'], figsize=(7, 5))

# 3. Add Percentage Labels automatically
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

# 4. Clean up labels
plt.title('Churn Rate: Senior Citizens vs Non-Seniors', fontsize=14)
plt.xticks(ticks=[0, 1], labels=['Non-Senior', 'Senior'], rotation=0)
plt.ylabel('Percentage (%)')
plt.legend(title='Churn', loc='upper right')
plt.tight_layout()
plt.show()
```

The number of churned is higher in senior citizen as compare to non-senior citizen .

```python
cols = ['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 
        'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 
        'Contract', 'PaperlessBilling', 'PaymentMethod']

for col in cols:
    plt.figure(figsize=(10, 5))
    ax = sns.countplot(x=col, hue='Churn', data=data, palette='viridis')
    plt.title(f'Churn Distribution for {col}')
    
    total = len(data)
    for container in ax.containers:
        labels = [f'{int(v.get_height())}\n({(v.get_height()/total)*100:.1f}%)' for v in container]
        ax.bar_label(container, labels=labels, label_type='edge', padding=3)
    
    plt.xticks(rotation='vertical')
    plt.show()
```

```python
for col in cols:
    plt.figure(figsize=(8, 8))
    
    # Group by the feature and Churn to get the breakdown
    chart_data = data.groupby([col, 'Churn']).size()
    
    chart_data.plot(kind='pie', autopct='%1.1f%%', startangle=140, 
                    cmap='Pastel1', wedgeprops={'edgecolor': 'black'})
    
    plt.title(f'Global % Breakdown: {col} & Churn')
    plt.ylabel('') # Remove the column name from the Y-axis
    plt.show()
```
